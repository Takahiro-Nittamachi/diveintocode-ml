### 標準偏差について、分散との違いや関係性、どんな時に使う指標なのか？

- 標準偏差とは
    - データの散らばり具合を表す値である。
    
- 標準偏差の求め方
    - 分散の正の平方根で求められる。
    - 分散とはそれぞれの値 $x_i$ から平均値 $\overline{x}$ を引いた値を2乗した値を求めた後、その値の平均値を取ることで求められる。
  
    分散$={\frac{1}{n}\sum_{n=1}^{n}{(x_i-\overline{x})}^2}$

    標準偏差$=\sqrt{\frac{1}{n}\sum_{n=1}^{n}{(x_i-\overline{x})}^2}$
        
        
- 標準偏差と分散の関係性
    - 値の相関関係
        - 分散の値が大きいと標準偏差の値も大きくなる。
    - 値の大小関係
        - 常に分散＞標準偏差の関係が成り立つ。※分散の正の正方根が標準偏差の為。

        
- 使用用途
    - 連続または非連続なデータを複数取得した場合、どのデータが最もバラツキがあるのか等を確認する際に標準偏差を使用する。

### 共分散行列について述べよ。

#### 共分散行列とは
- 分散共分散行列とも呼ばれ、正規分布の推定を行う際に必要となる行列である。
    
- 共分散行列の求め方
    - 確率変数$X = (X_1,X_2,\cdots,X_n)$という$X_1$から$X_n$までの確率変数をまとめたベクトルがあるとする。$X$の分散を$V[X]$、$X,Y$の共分散を$Cov[X,Y]$とした時、$V[\boldsymbol{X}]$を$X$の共分散行列とすると以下で求められる。
        
$$V[\boldsymbol{X}] = \left[\begin{array}{cccc}
          V[X_1] & Cov[{X_1,X_2}] & \cdots & Cov[{X_1,X_n}] \\
          Cov[{X_1,X_2}] & V[X_2] & \cdots & Cov[{X_2,X_n}]\\
          \vdots & \vdots & \ddots & \vdots \\
          Cov[{X_n,X_1}] & Cov[{X_n,X_2}] & \cdots & V[X_n]\\
        \end{array}\right]
$$

#### 共分散行列の値の意味
- 対角項(左上から右下方向)は各次元の乱数の分散に、非対角項(右上から左下方向)は$Cov[{X_1,X_2}]$に収束する。

$$V[\boldsymbol{X}] = \left[\begin{array}{cc}
          V[X_1] & Cov[{X_1,X_2}]\\
          Cov[{X_n,X_1}] & V[X_2]\\
        \end{array}\right]
$$

#### 共分散行列の具体的な使用用途「正規分布の推定」
- 例として以下の共分散行列 $V[\boldsymbol{X}]$ から確率変数 $X_1, X_2$の2次元正規分布を推定する場合を考える。

$$V[\boldsymbol{X}] = \left[\begin{array}{cc}
            5 & 4\\
            4 & 5\\
        \end{array}\right]
$$

1.固有値 $\lambda$ と固有ベクトル $r$ を求める為、以下の等式を計算する。

$$V[\boldsymbol{X}]r = \lambda r$$

 上記の等式を計算し、$\lambda = 1$の時の固有ベクトル $r$は${[1, -1]}^T$、$\lambda = 9$の時の固有ベクトル $r$は${[1, 1]}^T$となる。
    
2.求めた固有ベクトルの数値は、それぞれ正規分布の長軸方向のを表すベクトル($\lambda=9$の固有ベクトル)、短軸方向をあらわすベクトル($\lambda=1$の固有ベクトル)となる。
    
##### 上記①②の計算を行う事によって、実際の2次元正規分布が確認できない状態であっても、共分散行列から固有値と固有ベクトルを計算することによって、$X_1, X_2$の2次元正規分布の形を確認できる。
    
    
#### 共分散行列と機会学習の関係性
- 上記の「共分散行列を用いた正規分布の推定」は、機会学習における主成分分析(principal component analysis; PCA)を行う際の主成分を見つける事と同意である。
    


### 1次元の正規分布と多次元の正規分布について述べてください。特にどういう時に使うのか?

#### 1次元の正規分布の使用用途
例 あるクラスでテストを行い特定の点数以上である確率を求める場合。

#### 多次元の正規分布の使用用途

## 【課題】データ点の作成

2次元正規分布を用いて、中心(平均)が(-3,0)と(0,-3)、共分散行列が以下で表現される確率分布から生成されるデータ点を500個ずつ作成せよ。
その際、片方の500個のデータにはdata1、もう片方の500個のデータにはdata2と名付けて作成せよ。

$$
    \left[\begin{array}{cc}
          1.0 & 0.8 \\
          0.8 & 1.0 \\
    \end{array}\right]
$$

#### ヒント
np.random.multivariate_normalを用いて構いません。公式ドキュメントや日本語の記事にも情報が載っています。
2つの確率分布から生成されるデータは異なる変数で定義してください。

```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline

# データ点を格納するリストdata1、data2
data1 = []
data2 = []

# 平均
av1 = [-3, 0]
av2 = [0, -3]

# 共分散
cov = [[1.0, 0.8], [0.8, 1.0]]

# multivariate_noraml関数で平均av1、共分散covのデータ数500の正規分布のデータを作成し
# x1, y1に作成
x1, y1 = np.random.multivariate_normal(av1, cov, 500).T

# multivariate_noraml関数で平均av2、共分散covのデータ数500の正規分布のデータを作成。
# x2, y2に作成
x2, y2 = np.random.multivariate_normal(av2, cov, 500).T

# zip関数にてx1, y1のタプル形式のデータをリストdata1に、
# x2, y2のタプル形式のデータをリストdata2に格納。
for data in zip(x1, y1):
    data1.append(data)

for data in zip(x2, y2):
    data2.append(data)

# 散布図をプロットする
plt.scatter(x1, y1, color='r', marker='x', label='$cov-av1$')
plt.scatter(x2, y2, color='b', marker='x', label='$cov-av2$')

# ラベルを設定
plt.xlabel('$x$',size=20)
plt.ylabel('$y$',size=20)

# 軸を設定
plt.axis([-7.5,7.5,-7.5,7.5],size=20)
# plt.grid(True)
plt.legend();
```

```python
# data1の値を表示
df1 = pd.DataFrame(data1,columns = ['x1','y1'])
df1.head()
```

```python
# data2の値を表示
df2 = pd.DataFrame(data2,columns = ['x2','y2'])
df2.head()
```

## 【課題】データ点へのラベル付け

実際に扱うデータは各データが何のグループに属しているのかラベル付けされていなければなりません。
今回は二つの確率分布からデータをサンプリングしていますので、各500個のデータに対して、2つの確率分布のうち、どちらの確率分布からサンプリングされたものなのか、ラベル付けする必要があります。

上で作成した500個×2のデータに対してラベル付けをしてください、割り当てるラベルは区別できれば何でもいいですが、ここでは簡単に0と1にしてください。

やり方はたくさんありますが、ヒントを示します。

- ２つの確率分布からサンプリングされたデータ(numpy配列)をnp.concatenateを用いて結合する。
- 結合した行列の上から500番目までは1を割り振り、501番目~1000番目に0を割り振る。割り振る際は、結合した配列の３列目にラベルの列を追加する。1、2列目にはすでに、各データのx座標とy座標が格納されています。

```python
# data1にラベル0を付与
L0 = np.zeros(500, dtype=int) # zerosメソッドにて500個の0の[]
df1['label'] = pd.DataFrame(L0)
df1.head()
```

```python
# data2にラベル0を付与
L1 = np.ones(500, dtype=int)
df2['label'] = pd.DataFrame(L1)
df2.head()
```

```python
df3 = pd.DataFrame(np.concatenate([df2, df1])) # data1とdata2を縦に結合。
df3.columns = ['x', 'y', 'label']
# df3 = df3.astype({'label':int})　# 構文エラーが発生するので、後日確認する。
df3
```
