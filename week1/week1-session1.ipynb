## 【課題】100日後の米粒の数

100日後には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

### 100日後の米粒の合計

```python
total_rice = 1
for i in range(100):
    total_rice = total_rice + 2 ** i
print('100日後の米粒の合計:{:,}粒'.format(total_rice))
```

### 1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表示

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

list_total_grains = [1] # 累計の米の数を格納するリスト
list_n_grains = [1] # その日にもらう米の数を格納するリスト

for i in range(1, 100):
    list_total_grains.append(list_total_grains[i-1] + 2 ** i) # 累計の米の数
    list_n_grains.append(2**i) # その日にもらう米の数

df = pd.DataFrame(
    {'Total rice' : list_total_grains, # 累計の米の数
     'Increse rice' : list_n_grains # もらう米の数
    },)
df.index = np.arange(1, len(df) + 1)
df = df.astype(float)
pd.options.display.precision = 40
df
```

```python
df.plot()
plt.xlabel('day') # 日数
plt.ylabel('Amount of rice'); # 米の量
```

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（初日に貰うリストの長さはdayに等しくなる）
    """
    list_total_grains = [0]
    list_n_grains = [1]
    for i in range(1, day):
        list_total_grains.append(list_total_grains[i-1] + 2 ** i)
        list_n_grains.append(2**i)
            
    return list_n_grains, list_total_grains

"""機会学習エンジニアコースの受講期間は122日間の為、
122日間、米をもらった場合の米の数を関数で求める。"""
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=122)

df = pd.DataFrame(
    {'Cumsum' : list_total_grains,
     'Number of rice to receive' : list_n_grains
    },)
df.index = np.arange(1, len(df) + 1)
df = df.astype(float)
pd.options.display.precision = 40
df
```

## 【課題】何人が何日生活できる米か

この米があれば何人が何日生活できるか計算する関数を作成してください。機械学習コースを受講する期間に貰える米粒で、受講している人数であれば、何日間生活できるかを出力してみましょう。
関数の最低限の入力は米粒の数(int)と人数(int)、出力は生活できる日数(int)です。関数名やコメントなどは分かりやすく考えて作成してください。
ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

### 以下の条件でプログラムを作成。
- 条件①1人1日3食全てでお茶碗1杯(大盛りではない)のお米を消費する。
- 条件②お茶碗1杯のお米は3250粒とする。

### 上記の条件を設定した理由
- 実際の食生活をプログラムに導入すると、複雑になりすぎてしまう為、毎日同じ量の米を消費するように設定した。
- 情報の正確性の観点から公益社団法人 米穀安定供給確保支援機構の以下のページを確認し、お茶碗1杯はおよそ3250粒とした。
[参照リンク](http://www.komenet.jp/_qa/chawanippai/chawan_ippai02.html)

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（初日に貰うリストの長さはdayに等しくなる）
    """
    list_total_grains = [0]
    list_n_grains = [1]
    for i in range(1, day):
        list_total_grains.append(list_total_grains[i-1] + 2 ** i)
        list_n_grains.append(2**i)
            
    return list_n_grains, list_total_grains


def compute_eating_period(rice_amount=10000, number_of_students=1):
    """
    引数で指定された米の数と人数より、１日３食お茶碗１杯のご飯を食べた場合、
    何日分になるのかを計算する。

    Parameteres
    --------------
    rice_amount: int
        米粒の数 (default : 10000)
    people: int
        人数

    Returns
    --------------
    period_year : int
        指定された米と人数で1日3食を食べ続けていける期間(年単位)
    period_day : int
        同上(余りの日数)
    """
    # 1年間の日数
    NUMBER_OF_DAYS_IN_YEAR = 365
    
    # お茶碗1杯の米の数
    NUMBER_OF_RICE_PER_CUP = 3250 
    
    # 指定された人数が1日3食お米を食べて消費する米の量
    amount_of_rice_per_day = (number_of_students * 3) * NUMBER_OF_RICE_PER_CUP
    
    # 食べ続けていける日数(～日間)
    base_period = rice_amount // amount_of_rice_per_day
    
    # 食べていける年数と日数を計算。
    period_year, period_day = divmod(base_period, NUMBER_OF_DAYS_IN_YEAR)
    
    return period_year, period_day

"""機会学習エンジニアコースの受講期間は122日間の為、
122日間、毎日、米をもらった場合の米の数を関数で求める。"""
period = 122
students = 13

list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=period)
period_year, period_day = compute_eating_period(rice_amount=list_total_grains[period - 1], number_of_students = students)

print('''{}日間、曽呂利新左衛門の米の逸話の通りお米を貯めた場合、
{}人が{}年{}日間食べていける'''.format(period,students, period_year, period_day))
```
